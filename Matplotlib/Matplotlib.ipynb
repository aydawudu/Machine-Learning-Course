# Matplotlib example workflow

```python
#0- import matplotlip
%matplotlib inline
import matplotlib.pyplot as plt

#1. prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#3.setup plot
fig, ax=plt.subplots(figsize=(5,7))#width and height

#3. plot the data
ax.plot(x,y)

#4.customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
       ylabel="y_axis")

#5. Save and show the figure
fig.savefig("../Matplotlib/sample-plot.png")
```

### Two options for subplots

```python
#subplots option 1
fig, ((ax1, ax2), (ax3, ax4))=plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#subplots option 2
fig, ax=plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,5))
#plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### choosing between pyplot and matplotlib: 
* When plotting something quickly use the pyplot method
* but when plotting something more advance use the OO method

```python
heart_health.head()
```

```python
over_50=heart_health[heart_health["age"]>50]
over_50
```

```python
len(over_50)
```

```python
#pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,7));
```

```python
#OO method mixed wiht pyplot
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);

#ax.set_xlim([45,100])
```

```python
#OO method from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"])

#customize
ax.set(title="Health health and chol levels",
      xlabel="Age",
      ylabel="Chol")

#Add legend
ax.legend(*scatter.legend_elements(), title="Target")

#add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
plt.style.use('seaborn')

#Subplots of chol, age, thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True)
#add data to axis 0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="summer")
#customize
ax0.set(title="Health check",
       ylabel="Chol")

#add legend
ax0.legend(*scatter.legend_elements(), title="Target")
#add meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");
#add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="winter")
#customize
ax1.set(title="Heart Check",
       xlabel="Age",
       ylabel="Thalach")
#add legend
ax1.legend(*scatter.legend_elements(), title="Target")

#add meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

#Add title to the figure
fig.suptitle("Health Check Analysis", fontsize=16, fontweight="bold");
```

```python
#Customerizing the y and x axis limitations

plt.style.use('seaborn')

#Subplots of chol, age, thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True)
#add data to axis 0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter")
#customize
ax0.set(title="Health check",
       ylabel="Chol")
ax0.set_xlim([50,80])

#add legend
ax0.legend(*scatter.legend_elements(), title="Target")
#add meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");
#add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="winter")
#customize
ax1.set(title="Heart Check",
       xlabel="Age",
       ylabel="Thalach")

#change ax1 x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#add legend
ax1.legend(*scatter.legend_elements(), title="Target")

#add meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

#Add title to the figure
fig.suptitle("Health Check Analysis", fontsize=16, fontweight="bold");
```
